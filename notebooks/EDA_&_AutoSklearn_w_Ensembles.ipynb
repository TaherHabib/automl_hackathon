### Experiments into converting Categorical Data fields to Numerical 

```python
def convertCategoricalsToNumerics(df):
    
    categorical_fields = df.select_dtypes(include=['object'])
    categorical_fields = categorical_fields.drop('class',axis=1)
    
    for col in df.columns:
        if col in categorical_fields.columns:
                       
            # Counting the different categories in the column
            countOrdered_cats = list(dict(categorical_fields[col].value_counts()).keys()) # descending order
            
            
            # Creating a list of numeric replacements
            num_replacements = list(np.arange(len(countOrdered_cats))+1)
            
            # Dictionary of replacements to pass to df.replace()
            replacements = dict(zip(countOrdered_cats, num_replacements))
    
            cd = df[col].replace(replacements, inplace=True)
        
        else:
            pass
        
    return df
    
```

```python
train_numData = convertCategoricalsToNumerics(train_data)
```

```python
X = train_numData.loc[:, train_numData.columns != 'class']

Y = train_numData.loc[:, train_numData.columns == 'class']
```

```python
import autosklearn.classification
import autosklearn.metrics

from sklearn.model_selection import train_test_split

test_size = 0.2
shuffle = True

train_X, test_X, train_Y, test_Y = train_test_split(X, Y, test_size=test_size, shuffle=shuffle)
```

## Building a Manual Ensemble using Sklearn

```python
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, MaxAbsScaler
from sklearn.model_selection import RandomizedSearchCV

from sklearn.experimental import enable_hist_gradient_boosting
from sklearn.ensemble import HistGradientBoostingClassifier


numeric_transformer = make_pipeline(MaxAbsScaler())

Preprocessor = ColumnTransformer(
 transformers=[
 ('numeric_transformer', numeric_transformer, train_X.select_dtypes(exclude=['object']).columns),
 ])

hgb_pipe = make_pipeline(Preprocessor, HistGradientBoostingClassifier())

search_space = {
    'histgradientboostingclassifier__learning_rate': [0.0001, 0.001, 0.01, 0.1, 0.2, 0.3],
    'histgradientboostingclassifier__max_leaf_nodes': [8, 16, 32, 64, 128, 256, 512, 1024],
    'histgradientboostingclassifier__min_samples_leaf': [10, 20, 50, 100, 200, 500, 800, 1000],
    'histgradientboostingclassifier__scoring': ['accuracy'],
    'histgradientboostingclassifier__loss': ['categorical_crossentropy'],
    'histgradientboostingclassifier__random_state': [53]
}
```

```python
#instantiate the Random CV Search
hgb_grid = RandomizedSearchCV(hgb_pipe, search_space, 
                              n_jobs=-1, 
                              cv=5, 
                              scoring='accuracy',
                              verbose=10, 
                              refit=True)

hgb_grid.fit(train_X, train_Y)
```

```python
print(hgb_grid.best_params_)
print(hgb_grid.best_score_)
```

## Using AutoSklearn to find Ensembles

```python
from ConfigSpace.configuration_space import ConfigurationSpace
from ConfigSpace.hyperparameters import UniformFloatHyperparameter, UniformIntegerHyperparameter, OrdinalHyperparameter

import sklearn.ensemble
import sklearn.tree

from autosklearn.pipeline.components.base import AutoSklearnClassificationAlgorithm
from autosklearn.pipeline.constants import DENSE, UNSIGNED_DATA, PREDICTIONS, SPARSE


class CenturionsClassifier(AutoSklearnClassificationAlgorithm):

    def __init__(self, learning_rate, 
                 max_leaf_nodes, max_depth,
                 min_samples_leaf,
                 loss='categorical_crossentropy',
                 random_state=None):
        
        self.loss = loss
        self.learning_rate = float(learning_rate)
        self.max_leaf_nodes = int(max_leaf_nodes)
        self.max_depth = int(max_depth)
        self.min_samples_leaf = int(min_samples_leaf)
        self.random_state = random_state
        self.scoring = 'accuracy'

    def fit(self, X, Y, sample_weight=None):
        estimator = sklearn.ensemble.HistGradientBoostingClassifier(
            loss=self.loss,
            learning_rate=self.learning_rate,
            max_leaf_nodes=self.max_leaf_nodes,
            max_depth=self.max_depth,
            min_samples_leaf=self.min_samples_leaf,
            scoring=self.scoring,
            random_state=self.random_state
        )

        estimator.fit(X, Y, sample_weight=sample_weight)
        self.estimator = estimator
        
        return self

    def predict(self, X):
        if self.estimator is None:
            raise NotImplementedError
        return self.estimator.predict(X)

    def predict_proba(self, X):
        if self.estimator is None:
            raise NotImplementedError()
        return self.estimator.predict_proba(X)
    
    @staticmethod
    def get_properties(dataset_properties=None):
        return {'shortname': 'CC',
                'name': 'Centurions Classifier',
                'handles_regression': False,
                'handles_classification': True,
                'handles_multiclass': True,
                'handles_multilabel': False,
                'handles_multioutput': False,
                'is_deterministic': True,
                'input': (DENSE, SPARSE, UNSIGNED_DATA),
                'output': (PREDICTIONS,)}

    @staticmethod
    def get_hyperparameter_search_space(dataset_properties=None):
        cs = ConfigurationSpace()

        learning_rate = OrdinalHyperparameter(
            name='learning_rate', sequence=[0.0001, 0.001, 0.01, 0.1, 0.2, 0.3])
        max_leaf_nodes = UniformIntegerHyperparameter(
            name="max_leaf_nodes", lower=8, upper=512, default_value=32, log=False)
        max_depth = UniformIntegerHyperparameter(
            name="max_depth", lower=2, upper=1024, default_value=2, log=False)
        min_samples_leaf = UniformIntegerHyperparameter(
            name="min_samples_leaf", lower=10, upper=1000, default_value=20, log=False)
        
        cs.add_hyperparameters([learning_rate, max_leaf_nodes, max_depth, min_samples_leaf])
        return cs
```

```python
from autosklearn.pipeline.components.classification import ClassifierChoice

autosklearn.pipeline.components.classification.add_classifier(CenturionsClassifier)

for name in ClassifierChoice.get_components():
    print(name)


# Create and train an ensemble with AutoML
estimator_askl_mod = autosklearn.classification.AutoSklearnClassifier(
    time_left_for_this_task=1000,
    seed=1, 
    resampling_strategy='cv',
    metric=autosklearn.metrics.accuracy,
    # It is possible to run Auto-sklearn in parallel with additional components 
    # if these components are in separate modules and the components are 
    # imported when using auto-sklearn 0.14.2. For this example we are using 
    # auto-sklearn 0.14.0 so it does not work yet
    n_jobs=-1,
    # above is same as before, but here we restrict the design space to your new model
    include={
        'classifier': [
            'CenturionsClassifier',
        ]
    },
)
# Auto-sklearn ingests the pandas dataframe and detects column types
estimator_askl_mod.fit(train_X, train_Y)
```

## Using AutoSklearn to Find a Pipeline

```python
import sklearn

def scoring_function(estimator, X, Y):
    
    predictions = estimator.predict(X)
    return sklearn.metrics.accuracy_score(Y, predictions)
```

```python
# Create and train an ensemble with AutoML 
estimator = autosklearn.classification.AutoSklearnClassifier(
    time_left_for_this_task=1000000, # in seconds
    seed=42, 
    resampling_strategy='cv',
    resampling_strategy_arguments= {'shuffle':True, 'folds': 5},
    metric=autosklearn.metrics.accuracy,
    n_jobs=8
)
```

```python
# Auto-sklearn ingests the pandas dataframe and detects column types
estimator.fit(train_X, train_Y, dataset_name='hackathon_data')
```

```python
print(f"Train Auto-Sklearn Classifier performance is {scoring_function(estimator, train_X, train_Y)}")
print(f"Test Auto-Sklearn Classifier performance is {scoring_function(estimator, test_X, test_Y)}")
```
